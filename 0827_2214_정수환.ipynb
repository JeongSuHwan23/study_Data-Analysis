```python
#딕셔러리 로그인 예제

from os import truncate

members = {'urkpo': '0928^7$',
           'xxayv': '%2*9$91',
           'lsqvx': '!0%)&&4',
           'heums': '%@3^0%3',
           'uwcmc': '85236(&',
           'iemwv': ')8!36^&',
           'sqblx': ')^2)9!(',
           'jbbpy': '67269*3',
           'hjkwu': '$&@@#64',
           'fvwwy': '82$%)31'}
flagId = 1
flagPassword = 1
while flagId or flagPassword:
  if flagId == 1:
    id = input('ID : ')
    if id in members:
      print('아이디 확인 완료')
      flagId = 0
    else:
      print('아이디를 다시 입력하세요')
      continue
  password = input('Password : ')
  if password == members.get(id):
    print('로그인 성공')
    flagPassword = 0
  else:
    print('비밀번호를 다시 입력하세요')



```

```python
print('주민등록번호 유효성 검사')

a = input().replace('-', '')


b = int(a[0]) * 2 + int(a[1]) * 3 + int(a[2]) * 4 + int(a[3]) * 5 + \
    int(a[4]) * 6 + int(a[5]) * 7 + int(a[6]) * 8 + int(a[7]) * 9 + \
    int(a[8]) * 2 + int(a[9]) * 3 + int(a[10]) * 4 + int(a[11]) * 5

c = b % 11
d = 11 - c

if d > 9:
    d = d - 10

if d == int(a[12]):
    print("유효한 주민등록번호입니다.")
else:
    print("유효하지 않는 주민등록번호입니다.")
```

```python
#분자식

atomic = {
    'H': 1,
    'C': 12,
    'O': 16,
    'N': 14
}

molecular = input("분자식을 입력하세요: ")

i = 0
weight = 0

while i < len(molecular):
    e = molecular[i]
    i += 1
    quantity = 0
    while i < len(molecular) and molecular[i].isdigit():
        quantity = quantity * 10 + int(molecular[i])
        i += 1
    if quantity == 0:
        quantity = 1
    weight += atomic[e] * quantity

print(weight)
```

```python
#프로젝트 : 전화번호

user = {}
flag = 1
while flag:
  print('1 : 전화번호 등록')
  print('2 : 전화번호 삭제')
  print('3 : 전화번호 검색')
  print('4 : 전화번호 전체 보기')
  print('5 : 종료')
  n = input()
  if(n == '1'):
    u = input('이름:전화번호 형태로 입력 >> ')
    user[u[0:3]] = u[4:]
  elif(n == '2'):
    u = input('이름 입력 : ')
    del(user[u])
  elif(n == '3'):
    u = input('이름 입력 : ')
    print(user.get(u))
  elif(n == '4'):
    print(user)
  elif(n == '5'):
    print('종료')
    flag = 0
```
